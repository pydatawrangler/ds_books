## Example = parsing a single tweet

```python
first_tweet = "RT @j_o_n_dnger: $TWTR now top holding for Andor, unseating $AAPL"
```

```python
words_in_tweet = first_tweet.split(' ')
```

```python
words_in_tweet
```

```python
for word in words_in_tweet:
    if "$" in word:
        print("THIS TWEET IS ABOUT", word[1:])
```
